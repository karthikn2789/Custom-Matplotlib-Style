## Data Visualization Example using Matplotlib

This is an example data visualization example using Matplotlib where I do not like any of the default styles. So, I have decide to customise the style for my graphs. The aspects or properties that I shall be customising are:  
1. Set a default figure size of 5 x 4 inches with white background
2. Use a four color cycle based on cyan, magenta, yellow and black
3. Remove top and right spines
4. Both axis tick customisations:
    1. Move tick marks to inside
    2. Increase major tick width to 2
    3. Increase tick label's size
5. Change title font to Arial and make it bigger
6. Change both axis label font to Arial and make it bigger

Let us make three graphs. We have to customise each graph individually. This becomes a repeated task of copying and pasting the customised styles onto every graph that we make from here on. This is the worst case scenario trying to customise matplotlib styles.

````python
import matplotlib.pyplot as plt
import numpy as np
import math
from cycler import cycler

# Displays figures next to the cell that plots the figure
%matplotlib inline 
````

````python
# Setting a custom colour cycle using cyan, magenta, yellow and black
custom_cycler = (cycler(color=['c', 'm', 'y', 'k']))
````

### Bar Plots

````python
fig = plt.figure(figsize=(5,4), facecolor='w')

ax = fig.add_axes([0,0,1,1])

ax.set_prop_cycle(custom_cycler)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_tick_params(direction='in', which='major', width=2, labelsize='medium')
ax.yaxis.set_tick_params(direction='in', which='major', width=2, labelsize='medium')

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]

ax.bar(langs,students)

ax.set_title('Programming Language Popularity', fontname='Arial', fontsize='x-large')
ax.set_xlabel('Programming Language', fontname='Arial', fontsize='large')
ax.set_ylabel('Usage (in millions)', fontname='Arial', fontsize='large')

plt.show()
````

````python
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]

X = np.arange(4)
width = 0.25

fig = plt.figure(figsize=(5,4), facecolor='w')

ax = fig.add_axes([0,0,1,1])

ax.set_prop_cycle(custom_cycler)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_tick_params(direction='in', which='major', width=2, labelsize='medium')
ax.yaxis.set_tick_params(direction='in', which='major', width=2, labelsize='medium')

ax.bar(X + 0.00, data[0], width = 0.25)
ax.bar(X + 0.25, data[1], width = 0.25)
ax.bar(X + 0.50, data[2], width = 0.25)

ax.set_xticks(X+width)
ax.set_xticklabels(('2015','2016','2017','2018'))

ax.set_title('Performance Comparison', fontname='Arial', fontsize='x-large')
ax.set_xlabel('Departments', fontname='Arial', fontsize='large')
ax.set_ylabel('Average', fontname='Arial', fontsize='large')

ax.legend(labels=["CS", "IT", "ECE"])

plt.show()
````

````python
N = 5

menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)

ind = np.arange(N) # the x locations for the groups
width = 0.35

fig = plt.figure(figsize=(5,4), facecolor='w')

ax = fig.add_axes([0,0,1,1])

ax.set_prop_cycle(custom_cycler)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_tick_params(direction='in', which='major', width=2, labelsize='medium')
ax.yaxis.set_tick_params(direction='in', which='major', width=2, labelsize='medium')

ax.bar(ind, menMeans, width)
ax.bar(ind, womenMeans, width,bottom=menMeans)

ax.set_xticks(ind)
ax.set_yticks(np.arange(0, 81, 10))

ax.set_title('Scores by Group and Gender', fontname='Arial', fontsize='x-large')
ax.set_xlabel('Groups', fontname='Arial', fontsize='large')
ax.set_ylabel('Scores', fontname='Arial', fontsize='large')

ax.legend(labels=['Men', 'Women'])

plt.show()
````

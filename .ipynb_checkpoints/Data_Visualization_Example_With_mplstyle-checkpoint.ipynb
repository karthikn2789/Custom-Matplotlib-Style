## Data Visualization Example using Matplotlib

This is the same example as in [Typical_Data_Visualization_Example](Typical_Data_Visualization_Example.ipynb). As I did not like any of the default styles, I had decided to customise the style for my graphs. The aspects or properties that I shall be customising are:  
1. Set a default figure size of 5 x 4 inches with white background
2. Use a four color cycle based on cyan, magenta, yellow and black
3. Remove top and right spines
4. Both axis tick customisations:
    1. Move tick marks to inside
    2. Increase major tick width to 2
    3. Increase tick label's size
5. Change title font to Arial and make it bigger
6. Change both axis label font to Arial and make it bigger

The only difference here is that I have made use of a matplotlib style sheet [demo_style.mplstyle](demo_style.mplstyle) file instead of copying and pasting the above mentioned customisations. The contents of the file looks like this:

```
# Set figure size to 5x4 inches with white background
figure.figsize:     5,4             # figure size in inches
figure.facecolor:   white           # figure facecolor

# Change to cyan, magenta, yellow and black color palette
axes.prop_cycle:   cycler('color', ['c', 'm', 'y', 'k'])

# Remove top and right spines
axes.spines.left:   True            # display axis spines
axes.spines.bottom: True
axes.spines.top:    False
axes.spines.right:  False

# Both axis tick customisations
xtick.direction:    in              # direction: {in, out, inout}
ytick.direction:    in              # direction: {in, out, inout}
xtick.major.width:  2               # major tick width in points
ytick.major.width:  2               # major tick width in points
xtick.labelsize:    medium          # fontsize of the tick labels
ytick.labelsize:    medium          # fontsize of the tick labels

# Change font and font size of title and both axis labels
font.family:        Arial
font.size:          14.0
axes.titlesize:     x-large         # fontsize of the axes title
axes.labelsize:     large           # fontsize of the x and y labels
```   

Let us make three graphs using the demo_style style sheet. The main advantage of using a style sheet is that you need worry about customising and concentrate on making graphs. All you need to do is add a single line `plt.style.use(<mplstyle file>)` to apply your customised style on all the graphs created in a notebook or script.

````python
import matplotlib.pyplot as plt
import numpy as np

# Displays figures next to the cell that plots the figure
%matplotlib inline 
````

````python
# Applying the custom style sheet to all of the graphs to be created in this notebook
plt.style.use('demo_style.mplstyle')
````

### Bar Plots

````python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]

ax.bar(langs,students)

ax.set_title('Programming Language Popularity')
ax.set_xlabel('Programming Language')
ax.set_ylabel('Usage (in millions)')

plt.show()
````

````python
data = [[30, 25, 50, 20],
[40, 23, 51, 17],
[35, 22, 45, 19]]

X = np.arange(4)
width = 0.25

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.bar(X + 0.00, data[0], width = 0.25)
ax.bar(X + 0.25, data[1], width = 0.25)
ax.bar(X + 0.50, data[2], width = 0.25)

ax.set_xticks(X+width)
ax.set_xticklabels(('2015','2016','2017','2018'))

ax.set_title('Performance Comparison')
ax.set_xlabel('Departments')
ax.set_ylabel('Average')

ax.legend(labels=["CS", "IT", "ECE"])

plt.show()
````

````python
N = 5

menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)

ind = np.arange(N) # the x locations for the groups
width = 0.35

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.bar(ind, menMeans, width)
ax.bar(ind, womenMeans, width,bottom=menMeans)

ax.set_xticks(ind)
ax.set_yticks(np.arange(0, 81, 10))

ax.set_title('Scores by Group and Gender')
ax.set_xlabel('Groups')
ax.set_ylabel('Scores')

ax.legend(labels=['Men', 'Women'])

plt.show()
````
